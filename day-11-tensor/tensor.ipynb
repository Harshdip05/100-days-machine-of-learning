```python
import numpy as np
```

#### 0D tensor/ scalar

```python
a = np.array(1)
print(a)
a
```

```python
print(a.shape)
a.ndim
```

#### 1d tensor / vector

```python
a = np.array([1,2,3,4])
print(a)
a
```

```python
a = np.array([[1,2,3],[4,5,6]])
print(a)
a
```

```python
print(a.shape)
a.ndim
```

#### nd tensor / tensor

```python
a = np.array([[[1,2,3],[4,5,6]]])
print(a)
```

```python
print(a.shape)
a.ndim
```
