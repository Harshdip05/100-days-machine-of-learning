pandas.read_excel() needs the openpyxl package for .xlsx files

```python
import pandas as pd
```

```python
pd.read_excel("Hospital Companies.xlsx",header=1)
```

| File Type  | Engine   |
| ---------- | -------- |
| `.csv`     | built-in |
| `.xlsx`    | openpyxl |
| `.xls`     | xlrd     |
| `.parquet` | pyarrow  |
