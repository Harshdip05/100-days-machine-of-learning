```python
import sqlite3

# Connect database
conn = sqlite3.connect("sample_csv.db")

print("Connected Successfully")

# Create cursor
cursor = conn.cursor()

# Create table
cursor.execute("""
CREATE TABLE IF NOT EXISTS placement (
    id INTEGER PRIMARY KEY,
    name TEXT,
    salary INTEGER
)
""")

# Insert sample data
cursor.execute(
    "INSERT INTO placement (id, name, salary) VALUES (?, ?, ?)",
    (1, "Alice", 70000)
)

cursor.execute(
    "INSERT INTO placement (id, name, salary) VALUES (?, ?, ?)",
    (2, "Bob", 50000)
)

# Save changes
conn.commit()

# Select data
cursor.execute("SELECT * FROM placement")

# Fetch all rows
result = cursor.fetchall()

# Print rows
for row in result:
    print(row)

# Close
cursor.close()
conn.close()
```
